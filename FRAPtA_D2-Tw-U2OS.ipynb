#### Aims:
    To test whether mitochondrial RNA granules fulfill criteria of formation by liquid-liquid phase sepparation.
    
#### H0:
    MRG-proteins do not exchange as rapidly.
    
#### Approach:
    1) Photobleach FASTKD2-eGFP foci in COS-7 and U2OS cells to measure fluorescence recovery.
    2) Photobleach TWINKLE-eGFP foci in COS-7 cells to measure fluorescence recovery as a reference.
    3) Inspect recovery curves

#### Use of script:
    1) Specify the location of the input data & an output-directory.
    Requirements:
       INPUT: .csv files of recoveries generated by TrackFRAP.
       Functions & classes from class-file in the same directory as this jupyter notebook.
    2) Run all the cells.
    3) Inspect Results & draw conclusions.
